Question: What are the locations with the fastest roller coasters (minimum of 10)?

```python
ax=df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean']\
    .plot(kind='barh',figsize=(10,8),title='Average RollCoaster Speed by Location',color=('blue'))
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
