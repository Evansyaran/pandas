#SUBSET DATA

```python
# Subsetting Columns
df = df[['title', 'description', 'publishTime',
         'duration_seconds', 'viewCount', 'likeCount',
         'commentCount']]
```

```python
df.head()
```

```python
df.shape
```

```python
# Subsetting using query
df.query( 'viewCount > 1000000')
```

```python
#Removing rows with Nan values
df=df.loc[~ df['viewCount'].isna()]
```

```python
df=df.loc[~ df['likeCount'].isna()]
```

#Casting dtypes

```python
df['viewCount']=df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
```

```python
df.dtypes
```

```python
df['publishTime']
```

```python
df['publishTime']=pd.to_datetime(df['publishTime'])
```

```python
#CASTING using pd.to_numeric
pd.to_numeric(df['likeCount'].astype('str'))
```
